### 테스트 파일명 set

```python
import os

path="./input"
file_list = os.listdir(path)

print ("file_list: {}".format(file_list))
```

```python
# 테스트 파일 개수 확인
print(len(file_list))
```

### 테스트 파일 명 변경

```python
for i in range(len(file_list)):
    src = os.path.join(path, file_list[i])
    dst = "ex"+str(i+1)+".png"
    dst = os.path.join(path, dst)
    os.rename(src, dst)
```
